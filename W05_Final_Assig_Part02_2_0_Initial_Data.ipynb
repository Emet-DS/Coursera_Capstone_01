```python
# Import necesary libraries
import pandas as pd
import json # library to handle JSON files
import requests # library to handle requests
```

```python
# Get our target San Diego Zip codes that we processed and documented in a CSV file

San_Diego_codes = pd.read_csv ('SD_Metro_Zipcodes.csv', sep = ',', index_col=0)

San_Diego_codes.head()
```

```python
# Count the number of target Zip codes 
San_Diego_codes.shape
```

```python
# Test the API
# To get additional information from Zip-Codes.com
# API request url is as follows
# https://api.zip-codes.com/ZipCodesAPI.svc/1.0/GetZipCodeDetails/90210?key=DEMOAPIKEY

# create the API request URL
single_zip = 90210
api_key = 'DEMOAPIKEY'
zip_url = 'https://api.zip-codes.com/ZipCodesAPI.svc/1.0/GetZipCodeDetails/{}?key={}'.format(single_zip,api_key)

# make the GET request
get_zip_detail = requests.get(zip_url).json()['item']

```

```python
# Prepare to add information to a dataframe
sd_zip_details = pd.DataFrame(columns=['ZipCode', 
                                       'ZipCodePopulation', 'HouseholdsPerZipcode',
                                       'WhitePop', 'BlackPop','HispanicPop','AsianPop','IndianPop','HawaiianPop','OtherPop',
                                       'MalePop','FemalePop',
                                       'PersonsPerHousehold',
                                       'AverageHouseValue','IncomePerHousehold',
                                       'MedianAge','AverageFamilySize',
                                       'Latitude','Longitude',
                                       'AreaLand','AreaWater',
                                       'City','CountyName'])

```

```python
# Get data that is relevant for our analisys

record_append = []

record_append = [single_zip, 
            int(get_zip_detail['ZipCodePopulation']), 
            int(float(get_zip_detail['HouseholdsPerZipcode'])), 
            int(get_zip_detail['WhitePop']),
            int(get_zip_detail['BlackPop']),
            int(get_zip_detail['HispanicPop']),
            int(get_zip_detail['AsianPop']),
            int(get_zip_detail['IndianPop']),
            int(get_zip_detail['HawaiianPop']),
            int(get_zip_detail['OtherPop']),
            int(get_zip_detail['MalePop']),
            int(get_zip_detail['FemalePop']),
            float(get_zip_detail['PersonsPerHousehold']),
            float(get_zip_detail['AverageHouseValue']),
            float(get_zip_detail['IncomePerHousehold']),
            float(get_zip_detail['MedianAge']),
            float(get_zip_detail['AverageFamilySize']),
            float(get_zip_detail['Latitude']),
            float(get_zip_detail['Longitude']),
            float(get_zip_detail['AreaLand']),
            float(get_zip_detail['AreaWater']),
            get_zip_detail['City'],
            get_zip_detail['CountyName']]

df_length = len(sd_zip_details)
sd_zip_details.loc[df_length] = record_append
```

```python
# We iterate our API request for each of the 
# tested som test_zips = [91901, 91902, 91905] first
# for i in test_zips:

for i in San_Diego_codes['PostalCode']:
    # Print zipcode so we can review API progress
    print(i)
    zip_url = 'https://api.zip-codes.com/ZipCodesAPI.svc/1.0/GetZipCodeDetails/{}?key={}'.format(i,api_key)
    
    # make the GET request
    get_zip_detail = requests.get(zip_url).json()['item']
    
    # Populate the information into our DataFrame
    record_append = []

    record_append = [i, 
            int(get_zip_detail['ZipCodePopulation']), 
            int(float(get_zip_detail['HouseholdsPerZipcode'])), 
            int(get_zip_detail['WhitePop']),
            int(get_zip_detail['BlackPop']),
            int(get_zip_detail['HispanicPop']),
            int(get_zip_detail['AsianPop']),
            int(get_zip_detail['IndianPop']),
            int(get_zip_detail['HawaiianPop']),
            int(get_zip_detail['OtherPop']),
            int(get_zip_detail['MalePop']),
            int(get_zip_detail['FemalePop']),
            float(get_zip_detail['PersonsPerHousehold']),
            float(get_zip_detail['AverageHouseValue']),
            float(get_zip_detail['IncomePerHousehold']),
            float(get_zip_detail['MedianAge']),
            float(get_zip_detail['AverageFamilySize']),
            float(get_zip_detail['Latitude']),
            float(get_zip_detail['Longitude']),
            float(get_zip_detail['AreaLand']),
            float(get_zip_detail['AreaWater']),
            get_zip_detail['City'],
            get_zip_detail['CountyName']]

    df_length = len(sd_zip_details)
    sd_zip_details.loc[df_length] = record_append
```
